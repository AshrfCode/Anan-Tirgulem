```python
url = "https://data.gov.il/api/3/action/datastore_search?resource_id=053cea08-09bc-40ec-8f7a-156f0677aff3&limit=10000"

import requests
import pandas as pd
import ipywidgets as widgets
from ipywidgets import VBox, Output
from IPython.display import display

pd.set_option("display.max_columns", None)

response = requests.get(url)
data = response.json()

data_df = pd.DataFrame(data["result"]["records"])

output_area = Output()

tozeret_nm_dropdown = widgets.Dropdown(
    options=[""] + sorted(data_df["tozeret_nm"].dropna().unique().tolist()),
    description="Tozeret:",
    style={"description_width": "initial"}
)

kinuy_mishari_dropdown = widgets.Dropdown(
    options=[""],
    description="Kinuy Mishari:",
    style={"description_width": "initial"}
)

def update_kinuy_mishari_options(change):
    selected_tozeret = change["new"]

    if selected_tozeret:
        filtered_values = data_df[
            data_df["tozeret_nm"] == selected_tozeret
        ]["kinuy_mishari"].dropna().unique()

        kinuy_mishari_dropdown.options = [""] + sorted(filtered_values.tolist())
    else:
        kinuy_mishari_dropdown.options = [""]

    update_output()

def update_output(change=None):
    output_area.clear_output()

    selected_tozeret = tozeret_nm_dropdown.value
    selected_kinuy = kinuy_mishari_dropdown.value

    with output_area:
        if selected_tozeret and selected_kinuy:
            filtered_df = data_df[
                (data_df["tozeret_nm"] == selected_tozeret) &
                (data_df["kinuy_mishari"] == selected_kinuy)
            ]

            total_records = len(filtered_df)

            unique_ramat_gimur = (
                filtered_df["ramat_gimur"]
                .dropna()
                .unique()
                .tolist()
            )

            print(f"Total Records: {total_records}")

            if len(unique_ramat_gimur) > 0:
                print("Unique Ramat Gimur:")
                for ramat in unique_ramat_gimur:
                    print("-", ramat)
            else:
                print("Unique Ramat Gimur: None")
        else:
            print("Please select valid options for both dropdowns.")

tozeret_nm_dropdown.observe(update_kinuy_mishari_options, names="value")
kinuy_mishari_dropdown.observe(update_output, names="value")

display(VBox([tozeret_nm_dropdown, kinuy_mishari_dropdown, output_area]))
```

```python
import plotly.graph_objects as go
import datetime
import numpy as np

np.random.seed(1)

programmers = ['Alex','Nicole','Sara','Etienne','Chelsea','Jody','Marianne']

base = datetime.datetime.today()
dates = base - np.arange(180) * datetime.timedelta(days=1)

z = np.random.poisson(size=(len(programmers), len(dates)))

fig = go.Figure(data=go.Heatmap(
    z=z,
    x=dates,
    y=programmers,
    colorscale='Viridis'   # תיקון הגרש כאן
))

fig.update_layout(
    title=dict(text='GitHub commits per day'),
    xaxis_nticks=36
)

fig.show()
```

```python
import plotly.graph_objects as go

import pandas as pd

df1 = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/718417069ead87650b90472464c7565dc8c2cb1c/sunburst-coffee-flavors-complete.csv')
df2 = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/718417069ead87650b90472464c7565dc8c2cb1c/coffee-flavors.csv')

fig = go.Figure()

fig.add_trace(go.Sunburst(
    ids=df1.ids,
    labels=df1.labels,
    parents=df1.parents,
    domain=dict(column=0)
))

fig.add_trace(go.Sunburst(
    ids=df2.ids,
    labels=df2.labels,
    parents=df2.parents,
    domain=dict(column=1),
    maxdepth=2
))

fig.update_layout(
    grid= dict(columns=2, rows=1),
    margin = dict(t=0, l=0, r=0, b=0)
)

fig.show()
```

```python
import requests
import pandas as pd
import plotly.express as px
import ipywidgets as widgets

from IPython.display import display, clear_output
from ipywidgets import VBox, Tab, Output

url = "https://data.gov.il/api/3/action/datastore_search?resource_id=053cea08-09bc-40ec-8f7a-156f0677aff3&limit=50000"

response = requests.get(url)
data = response.json()

data_df = pd.DataFrame(data["result"]["records"])

stats_output = Output()
table_output = Output()
graphs_output = Output()
filter_result_output = Output()

# ================================
# Tab 1 - Statistics
# ================================
with stats_output:
    clear_output()

    print("Statistics of all data")
    print("----------------------")
    print("Number of records:", len(data_df))
    print("Number of columns:", len(data_df.columns))
    print("Number of manufacturers:", data_df["tozeret_nm"].nunique())
    print("Number of models:", data_df["kinuy_mishari"].nunique())

    if "shnat_yitzur" in data_df.columns:
        print("First year:", data_df["shnat_yitzur"].min())
        print("Last year:", data_df["shnat_yitzur"].max())

    display(data_df.describe(include="all"))

# ================================
# Tab 2 - Full Table
# ================================
with table_output:
    clear_output()
    display(data_df)

# ================================
# Tab 3 - Cars Per Year Graph
# ================================
with graphs_output:
    clear_output()

    cars_per_year = (
        data_df
        .groupby("shnat_yitzur")
        .size()
        .reset_index(name="count")
        .sort_values("shnat_yitzur")
    )

    fig = px.bar(
        cars_per_year,
        x="shnat_yitzur",
        y="count",
        title="Number of cars produced in each year",
        labels={
            "shnat_yitzur": "Year",
            "count": "Number of Cars"
        }
    )

    fig.show()

# ================================
# Tab 4 - Filter
# ================================
manufacturer_dropdown = widgets.Dropdown(
    options=[""] + sorted(data_df["tozeret_nm"].dropna().unique().tolist()),
    description="Manufacturer:",
    style={"description_width": "initial"}
)

model_dropdown = widgets.Dropdown(
    options=[""],
    description="Model:",
    style={"description_width": "initial"}
)

trim_dropdown = widgets.Dropdown(
    options=[""],
    description="Trim Level:",
    style={"description_width": "initial"}
)

def update_models(change):
    selected_manufacturer = manufacturer_dropdown.value

    if selected_manufacturer:
        models = (
            data_df[data_df["tozeret_nm"] == selected_manufacturer]["kinuy_mishari"]
            .dropna()
            .unique()
            .tolist()
        )
        model_dropdown.options = [""] + sorted(models)
    else:
        model_dropdown.options = [""]

    trim_dropdown.options = [""]
    update_filter_output()

def update_trim_levels(change):
    selected_manufacturer = manufacturer_dropdown.value
    selected_model = model_dropdown.value

    if selected_manufacturer and selected_model:
        trims = (
            data_df[
                (data_df["tozeret_nm"] == selected_manufacturer) &
                (data_df["kinuy_mishari"] == selected_model)
            ]["ramat_gimur"]
            .dropna()
            .unique()
            .tolist()
        )
        trim_dropdown.options = [""] + sorted(trims)
    else:
        trim_dropdown.options = [""]

    update_filter_output()

def update_filter_output(change=None):
    selected_manufacturer = manufacturer_dropdown.value
    selected_model = model_dropdown.value
    selected_trim = trim_dropdown.value

    filtered_df = data_df.copy()

    if selected_manufacturer:
        filtered_df = filtered_df[filtered_df["tozeret_nm"] == selected_manufacturer]

    if selected_model:
        filtered_df = filtered_df[filtered_df["kinuy_mishari"] == selected_model]

    if selected_trim:
        filtered_df = filtered_df[filtered_df["ramat_gimur"] == selected_trim]

    with filter_result_output:
        clear_output()

        if selected_manufacturer and selected_model:
            print("Manufacturer:", selected_manufacturer)
            print("Model:", selected_model)

            if selected_trim:
                print("Trim Level:", selected_trim)

            print("Number of cars:", len(filtered_df))

            print("\nAvailable trim levels:")
            trims_list = (
                data_df[
                    (data_df["tozeret_nm"] == selected_manufacturer) &
                    (data_df["kinuy_mishari"] == selected_model)
                ]["ramat_gimur"]
                .dropna()
                .unique()
                .tolist()
            )

            for trim in sorted(trims_list):
                print("-", trim)

            display(filtered_df)
        else:
            print("Please select manufacturer and model.")

manufacturer_dropdown.observe(update_models, names="value")
model_dropdown.observe(update_trim_levels, names="value")
trim_dropdown.observe(update_filter_output, names="value")

filter_tab_content = VBox([
    manufacturer_dropdown,
    model_dropdown,
    trim_dropdown,
    filter_result_output
])

# ================================
# Create Tabs
# ================================
tabs = Tab()

tabs.children = [
    stats_output,
    table_output,
    graphs_output,
    filter_tab_content
]

tabs.set_title(0, "Statistics")
tabs.set_title(1, "Full Table")
tabs.set_title(2, "Cars Per Year")
tabs.set_title(3, "Filter")

display(tabs)
```
