```python
import json
import os
from google.colab import drive
import ipywidgets as widgets
from IPython.display import display, clear_output

# ---------------------------
# Google Drive
# ---------------------------
drive.mount('/content/drive')

FILE_PATH = "/content/drive/MyDrive/students.json"

# ---------------------------
# Load JSON
# ---------------------------
def load_data():
    if not os.path.exists(FILE_PATH):
        print("File not found:", FILE_PATH)
        return []
    with open(FILE_PATH, "r", encoding="utf-8") as f:
        return json.load(f)

students = load_data()

if not students:
    print("No data found")
else:

    # ---------------------------
    # UI
    # ---------------------------
    title = widgets.HTML("<h2>🎓 Student System</h2>")

    emails = [s["email"] for s in students]

    dropdown = widgets.Dropdown(
        options=emails,
        description="בחר סטודנט:"
    )

    fname_box = widgets.Text(description="שם פרטי:", disabled=True)
    lname_box = widgets.Text(description="שם משפחה:", disabled=True)
    email_box = widgets.Text(description="אימייל:", disabled=True)
    courses_box = widgets.Text(description="קורסים:", disabled=True)
    favorite_box = widgets.Text(description="Favorite Link:")

    output = widgets.Output()

    # ---------------------------
    # פונקציות
    # ---------------------------
    def populate(email):
        index = emails.index(email)
        s = students[index]

        fname_box.value = s.get("first_name", "")
        lname_box.value = s.get("last_name", "")
        email_box.value = s.get("email", "")

        courses = s.get("courses", [])
        courses_box.value = ", ".join(courses) if isinstance(courses, list) else str(courses)

        favorite_box.value = s.get("favorite_link", "")

    def on_change(change):
        if change["type"] == "change" and change["name"] == "value":
            populate(change["new"])

    dropdown.observe(on_change)

    def save_update(b):
        with output:
            clear_output()

            index = emails.index(dropdown.value)
            students[index]["favorite_link"] = favorite_box.value

            with open(FILE_PATH, "w", encoding="utf-8") as f:
                json.dump(students, f, indent=4, ensure_ascii=False)

            print("✅ נשמר בהצלחה!")

    button = widgets.Button(description="עדכן", button_style="success")
    button.on_click(save_update)

    # ---------------------------
    # init
    # ---------------------------
    populate(emails[0])

    # ---------------------------
    # display
    # ---------------------------
    display(title)
    display(dropdown, fname_box, lname_box, email_box, courses_box, favorite_box, button, output)
```
