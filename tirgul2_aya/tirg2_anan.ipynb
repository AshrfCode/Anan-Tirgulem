```python
# חיבור ל-Google Drive
from google.colab import drive
drive.mount('/content/drive')

# קריאת קובץ JSON
import json

file_path = "/content/drive/MyDrive/students.json"

with open(file_path, 'r') as f:
    students = json.load(f)

# יצירת רשימת שמות
names = [s["name"] for s in students]

# ---------------------------
# Widgets
# ---------------------------
import ipywidgets as widgets
from IPython.display import display, clear_output

# Dropdown
dropdown = widgets.Dropdown(
    options=names,
    description='בחר סטודנט:'
)

# Text fields (חלק לקריאה בלבד)
name_box = widgets.Text(description="שם:", disabled=True)
email_box = widgets.Text(description="מייל:", disabled=True)
courses_box = widgets.Text(description="קורסים:", disabled=True)
favorite_box = widgets.Text(description="תוכנית אהובה:")

# Output area (להודעות)
output = widgets.Output()

# פונקציה לעדכון תצוגה
def update_display(selected_name):
    index = names.index(selected_name)
    student = students[index]

    name_box.value = student["name"]
    email_box.value = student["email"]
    courses_box.value = ", ".join(student["course"])
    favorite_box.value = student.get("favorite_program", "")

# מאזין לשינוי ב-dropdown
def on_change(change):
    if change["type"] == "change" and change["name"] == "value":
        update_display(change["new"])

dropdown.observe(on_change)

# כפתור עדכון
button = widgets.Button(description="עדכן")

def update_data(b):
    with output:
        clear_output()
        index = names.index(dropdown.value)
        students[index]["favorite_program"] = favorite_box.value

        # שמירה חזרה ל-JSON
        with open(file_path, "w") as f:
            json.dump(students, f, indent=4)

        print("עודכן בהצלחה ✅")

button.on_click(update_data)

# טעינה ראשונית (שלא יהיה ריק)
update_display(names[0])

# הצגה למסך
display(dropdown)
display(name_box, email_box, courses_box, favorite_box, button, output)
```
