```python
import json
file_path="/content/drive/MyDrive/students.json"
with open(file_path,'r') as f:
  students=json.load(f)
```

```python
names = [s["name"] for s in students]
import ipywidgets as widgets
from IPython.display import display

dropdown = widgets.Dropdown(
    options=names,
    description='בחר סטודנט:'
)

name_box = widgets.Text(description="שם:")
email_box = widgets.Text(description="מייל:")
courses_box = widgets.Text(description="קורסים:")
favorite_box = widgets.Text(description="תוכנית אהובה:")
```

```python
def on_change(change):
    index = names.index(change["new"])
    student = students[index]

    name_box.value = student["name"]
    email_box.value = student["email"]
    courses_box.value = ", ".join(student["course"])
    favorite_box.value = student.get("favorite_program", "")

dropdown.observe(on_change, names='value')
```

```python
button = widgets.Button(description="עדכן")

def update_data(b):
    index = names.index(dropdown.value)
    students[index]["favorite_program"] = favorite_box.value

    with open(file_path, "w") as f:
        json.dump(students, f, indent=4)

    print("עודכן בהצלחה ✅")

button.on_click(update_data)
```

```python
display(dropdown)
display(name_box, email_box, courses_box, favorite_box, button)
```
